```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#target value
true_value=50
```

```python
#simulate data
#1 accurate and precise(close to true value and tightly grouped)
'''
loc=true_value(true_value=50)the value will be
centered around the true value(50)
scale=1 the std deviation(spread) is small,
meaning the value will be tightly grped around the true value
this implies the higher precision
the measurement will vary only a little from true value
so they have be both accuracy(close to 50) and precise(close to each other)
'''
accurate_precise=np.random.normal(loc=true_value,scale=1,size=10)
```

```python
#2Accurate but not Precise(close to true value but spread out)
accurate_not_precise=np.random.normal(loc=true_value,scale=10,size=10)
```

```python
'''
The two types of code you have highlighted may look similar but they differ in one important aspect:
the value os scale which controls the spread of the generated values around the true value(loc)'''
```

```python
#3Precise but not Accurate(far from true value but tightly gropued)
precise_not_accurate=np.random.normal(loc=70,scale=1,size=10)
```

```python
#4Neither accurate nor precise(far from true value and spread out)
not_accurate_not_precise=np.random.normal(loc=70,scale=10,size=10)
```

```python
#Plot the results
plt.figure(figsize=(10,6))
```

```python
#Plot1:Accurate and precise
plt.scatter(accurate_precise,[1]*10,color='green',label='Accurate and Precise')

#Plot2:Accurate and not precise
plt.scatter(accurate_not_precise,[2]*10,color='blue',label='Accurate and Not Precise')

#Plot3:Precise but not Accurate
plt.scatter(precise_not_accurate,[3]*10,color='orange',label='Precise and Not Accurate')

#Plot4:Not Accurate Not Precise
plt.scatter(not_accurate_not_precise,[4]*10,color='red',label='Not Accurate and Not Precise')

#Adding target line
plt.axvline(true_value,color='black',linestyle='--',label='True value')

#Labels and legend
plt.yticks([1,2,3,4],['Accurate and Precise','Accurate but Not Precise','Precise but Not Accurate','Neither Accurate NOr Precise'])
plt.xlabel('Measurement value')
plt.legend()
plt.title('Accurate and Precision Demonstration')
```

Accurate=Close to true value i.e 50

Precise=Group together,std less

Accurate and Precise=mean near to 50,close to 50,std less

Accurate but Not Precise=mean near to 50,close to 50,std more

Precise but Not Accurate=mean greater than 50,not close to 50,std less

Neither Accurate NOr Precise=mean greater to 50,not close to 50,std more
