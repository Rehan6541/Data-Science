**#Visualizing time series Data**

```python
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_csv("starbucks.csv",index_col="Date",parse_dates=True)
```

```python
df.head()
```

```python
#to show that dates are already parsed
df.index
```

```python
df.plot()
```

```python
df['Close'].plot()
```

```python
title='Starbucks Closing Stocks Prices'
ylabel="Closing Price (USD)"
xlabel='Closing Date'

ax=df['Close'].plot(figsize=(12,6),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

**Choosing X limit by slice**

```python
#Date are seperated by a column 
df['Close']['2017-01-01':'2017-03-01'].plot(figsize=(12,4)).autoscale(axis='x',tight=True)
```

**Choosing X limit by argument**

```python
#Date are seperated by a comma
df['Close'].plot(figsize=(12,4),xlim=['2017-01-01','2017-03-01'])
```

```python
#Find the mininmum value in the range
df.loc['2017-01-01':'2017-03-01']['Close'].min()
```

```python
#Find the maximum value in the range
df.loc['2017-01-01':'2017-03-01']['Close'].max()
```

```python
#plud these in as y limit values:
df['Close'].plot(figsize=(12,4),xlim=['2017-01-01','2017-03-01'],ylim=[51.6899,56.3244])
```

```python
df['Close'].plot(figsize=(12,4),xlim=['2017-01-01','2017-03-01'],ylim=[0,60])
```
